````python
import json

def generate_product_brief(product_name, specs):

    prompt = f"""You are an expert eCommerce content strategist.

Given the following product information, generate a complete, structured product brief.

Product Name: {product_name}
Specifications: {specs}

Return ONLY a valid JSON object with exactly these keys:
{{
  "title": "SEO-optimised product title, maximum 150 characters, includes brand-relevant keywords",
  "description": "A 150-word benefit-led product description written in second person, active voice. Translate specs into customer benefits. No generic openers.",
  "bullet_points": [
    "FEATURE NAME: Benefit explanation in 20-30 words.",
    "FEATURE NAME: Benefit explanation in 20-30 words.",
    "FEATURE NAME: Benefit explanation in 20-30 words.",
    "FEATURE NAME: Benefit explanation in 20-30 words.",
    "FEATURE NAME: Benefit explanation in 20-30 words."
  ],
  "category": "Top-level category → Subcategory → Product type",
  "keywords": [
    "keyword 1", "keyword 2", "keyword 3", "keyword 4", "keyword 5",
    "keyword 6", "keyword 7", "keyword 8", "keyword 9", "keyword 10"
  ]
}}

Return only the JSON. No preamble, no exlanation, no markdown formatting."""

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7,
        max_tokens=1500
    )

    raw = response.choices[0].message.content.strip()

    # Clean any accidental markdown formatting
    raw = raw.replace("```json", "").replace("```", "").strip()

    return json.loads(raw)

print("✅ Generator function ready")
````

````python
def print_product_brief(brief, product_name):

    print("=" * 60)
    print(f"PRODUCT BRIEF: {product_name.upper()}")
    print("=" * 60)

    print("\n📌 TITLE")
    print(f"   {brief['title']}")

    print("\n📝 DESCRIPTION")
    print(f"   {brief['description']}")

    print("\n✅ BULLET POINTS")
    for i, bullet in enumerate(brief['bullet_points'], 1):
        print(f"   {i}. {bullet}")

    print("\n📂 CATEGORY")
    print(f"   {brief['category']}")

    print("\n🔍 SEARCH KEYWORDS")
    keywords = brief['keywords']
    print(f"   {', '.join(keywords)}")

    print("\n" + "=" * 60)
    print("✅ Product brief generated successfully")
    print("=" * 60)

print("✅ Formatter ready")
````

````python
import os
from datetime import datetime

def save_brief_to_file(brief, product_name):
    """Save the product brief as a text file."""

    # Create output folder if it doesn't exist
    os.makedirs("product_briefs", exist_ok=True)

    # Create filename from product name and timestamp
    safe_name = product_name.replace(" ", "_").lower()
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    filename = f"product_briefs/{safe_name}_{timestamp}.txt"

    # Format the content
    content = f"""
============================================================
PRODUCT BRIEF: {product_name.upper()}
Generated: {datetime.now().strftime("%d %B %Y, %H:%M")}
============================================================

TITLE:
{brief.get('title', 'N/A')}

DESCRIPTION:
{brief.get('description', 'N/A')}

BULLET POINTS:
"""
    bullets = brief.get('bullet_points', [])
    for i, bullet in enumerate(bullets, 1):
        content += f"{i}. {bullet}\n"

    content += f"""
CATEGORY:
{brief.get('category', 'N/A')}

SEARCH KEYWORDS:
{', '.join(brief.get('keywords', []))}

============================================================
"""

    # Write to file
    with open(filename, 'w', encoding='utf-8') as f:
        f.write(content)

    print(f"✅ Brief saved to: {filename}")
    return filename

print("✅ Save function ready")
````
